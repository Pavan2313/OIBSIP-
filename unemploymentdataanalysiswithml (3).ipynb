```python
df = pd.read_csv('/kaggle/input/unemployment-in-india/Unemployment in India.csv')
df = pd.read_csv('/kaggle/input/unemployment-in-india/Unemployment_Rate_upto_11_2020.csv')
df.head()
```

```python
sns.pairplot(df)
```

```python
plt.bar(df['Region'],df['Estimated Unemployed Rate'])
plt.xticks(rotation = 90)
plt.xlabel('States')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment according to states')
plt.show()
```

```python
sns.histplot(x = 'Estimated Unemployed Rate',data = df, kde=True)
plt.title('Unemployment according to area')
plt.xlabel('Unemployment Rate')
plt.show()
```

```python
sns.lineplot(y = 'Estimated Unemployed Rate',x = 'Date', data=df)
plt.title('Unemployment according to area')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
sns.scatterplot(df['Region'])
plt.title('Unemployment according to region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.bar(df['Region'],df['Estimated Labour Participation Rate'])
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Labour Participation Rate')
plt.title('Labour Psrticipation according states')
plt.show()
```

```python
sns.histplot(x = 'Estimated Labour Participation Rate',data=df)
plt.title('Labour Participation according area')
plt.xlabel('Labour Participation Rate')
plt.show()
```

```python
sns.lineplot(y = 'Estimated Labour Participation Rate',x = 'Date', data=df)
plt.title('Labour participation according to area')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Laber Participation Rate')
plt.show()
```

```python
sns.scatterplot(df['Region'])
plt.title('Labour Participation Rate according to region')
plt.xlabel('Region')
plt.ylabel('Labe Participation Rate')
plt.show()
```

```python
unemployment = df[['Region','Estimated Unemployed Rate','States']]
figure = px.sunburst(unemployment,path=['States','Region'],
                     width=700,height=700,
                     title='Indian Unemployed Rate')
figure.show()
```
